```python
def filt(points, err_factor: float) -> bool:
    mean = np.absolute(points[0] - points[-1])
    center_diff = np.absolute(points[1] - mean)
    return center_diff > mean * err_factor
```

```python
errors = []
start = 0
while start < len(data) - 3:
    if filt(data[start:start+3], 5):
        errors.append(start + 1)
        start += 2
    else:
        start += 1

plt.plot(data, color='blue', marker='o')      
plt.plot(errors, data[errors], 'ro')
```
