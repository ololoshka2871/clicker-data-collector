# Тест экспоненты

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize
```

```python
def f(x: float, A: float, B: float) -> float:
    return A * (1 - np.exp(-B * x))
```

```python
A = 1
B = 0.01

x = np.linspace(0, 200, 20)
y = [f(x, A, B) for x in x]
```

[Тут](https://stackoverflow.com/a/3433503) Апроксимация любой функцией

```python
coeffs, _ = optimize.curve_fit(f, x, y, method='trf')
a = coeffs[0]
b = coeffs[1]
print(f'a = {a} (асимптота), b = {b} (скорость роста)')

y_a = [f(x, a, b) for x in x]

plt.plot(x, y, '.', x, y_a, '-')
```
